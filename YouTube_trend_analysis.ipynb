```python
import pandas as pd

df = pd.read_csv("trending_videos.csv")
df
```

```python
#inspect out data
df.info()
```

```python
df.isnull().sum()
```

```python
#check for duplicates
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
# display data types
df.dtypes
```

```python
# fill missing descriptions with "No description"
df['description'].fillna("NO Description",inplace=True)

# convert `published_at` to datetime
df['published_at'] = pd.to_datetime(df['published_at'])

# convert tags from string representation of list to actual list
df['tags'] = df['tags'].apply(lambda x:eval(x) if isinstance(x,str) else x)
```

```python
# descriptive statistics
df[['view_count','like_count','dislike_count','comment_count']].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
#subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# view count distribution
sns.histplot(df['view_count'],bins=30,kde=True,ax=axes[0], color='blue')
axes[0].set_title('View Count Distribution')
axes[0].set_xlabel('View Count')
axes[0].set_ylabel('Frequency')

# like count distribution
sns.histplot(df['like_count'],bins=30,kde=True,ax=axes[1],color='orange')
axes[1].set_title("Like Count Distribution")
axes[1].set_xlabel('like Count')
axes[1].set_ylabel('Frequency')

# comment count distribution
sns.histplot(df['comment_count'],bins=30,kde=True,color="red",ax=axes[2])
axes[2].set_title('Comment Count Distribution')
axes[2].set_xlabel('Comment Count')
axes[2].set_ylabel('Frequency')
```

```python
correlation=df[['view_count','like_count','comment_count']].corr()
```

```python
correlation=df[['view_count','like_count','comment_count']].corr()
sns.heatmap(correlation,center=0.34,annot=True, cmap='coolwarm', linewidths=0.5, linecolor='black')
```

```python
sns.countplot(y='category_name',order=df['category_name'].value_counts().index,palette='dark',data=df)
```

```python
# average engagement metrics by category
category_engagement = df.groupby('category_name')[['view_count','like_count','comment_count']].mean().sort_values(by='view_count',ascending=False)
sns.set(style="whitegrid")

#subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# view count by category
sns.barplot(y=category_engagement.index,x=category_engagement['view_count'],ax=axes[0])
axes[0].set_title("Average View Count by Category")

# like count by category
sns.barplot(y=category_engagement.index,x=category_engagement['like_count'],ax=axes[1])
axes[1].set_title("Average Like Count by Category")

# comment count by category
sns.barplot(y=category_engagement.index,x=category_engagement['comment_count'],ax=axes[2])
axes[2].set_title("Average Comment Count by Category")

plt.tight_layout()
```

```python
!pip install isodate
import isodate
# convert ISO 8601 duration to seconds
df['duration_sec']=df['duration'].apply(lambda x:isodate.parse_duration(x).total_seconds())

# convert seconds to range
df['duration_range']=pd.cut(df['duration_sec'],bins=[0,300,600,1200,3600,7200],labels=['0-5 min', '5-10 min', '10-20 min', '20-60 min', '60-120 min'])
```

```python
plt.figure(figsize=(10, 6))
# scatter plot for video length vs view count
sns.scatterplot(x=df['duration_sec'],y=df['view_count'],alpha=0.6,color='purple')
plt.title('Video Length vs View Count')
plt.xlabel('Video Length (seconds)')
plt.ylabel('View Count')
plt.show()
```

```python
# calculate the number of tags for each video
df['tag_count']=df['tags'].apply(len)

# scatter plot for number of tags vs view count
sns.scatterplot(x='tag_count',y='view_count',data=df)
```

```python
# extract hour of publication
df['hour']=df['published_at'].dt.hour

# bar chart for publish hour distribution
sns.countplot(x=df['hour'],palette='coolwarm')
```

```python
sns.scatterplot(y='view_count',x='hour',data=df)
```
